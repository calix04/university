```python
import glob
import os
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('서울시 대학 및 전문대학 DB.csv', encoding= 'cp949') # 서울시 대학 및 전문대학 DB csv 파일을 불러옴
df
```

```python
# dataset에 컬럼이 학교종류 학교명 설립 주소 행정시 등등인 것들을 입력한다
dataset = df[[
    '학교종류','학교명','설립','주소','행정시','행정구','행정동','전화번호','팩스번호','홈페이지','위도','경도'
]] 
dataset.head()
```

```python
# dataset에서 컬럼인 설립이 사립, 공립, 국립, 특별법국립, 국립대법인 중에서 하나라도 같을시 불러와서 df_un에 입력
df_un = dataset[(dataset['설립']=='사립')|(dataset['설립']=='공립')|(dataset['설립']=='국립')|(dataset['설립']=='특별법국립')
                |(dataset['설립']=='국립대법인')]
df_un.index = range(len(df_un))
print('서울시 내 대학교 수 :', len(df_un)) # 서울시 내 대학교 수가 나타나도록 설정
df_un.head()
```

```python
df_seoul_un = df_un[df_un['학교명'].str.contains('대학교')]
df_seoul_un.index = range(len(df_seoul_un)) #인덱스
df_seoul_un.head()
```

```python
# 각 구마다 대학교가 몇 곳씩 있는지 확인하는 코드
un_region = df_un.groupby('행정구')['학교명'].count().to_frame().sort_values(by='학교명', ascending=False)
un_region = un_region.reset_index()
un_region = un_region.set_index('행정구')
un_region
```

```python
geo_path = '02. skorea_municipalities_geo_simple.json' 
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
loc = [37.459759, 126.951115] # 위도(N), 경도(E)
data_size = len(df_seoul_un)

map_university = folium.Map(location=loc, # 좌표 대입
                            zoom_start=10) # 얼만큼의 확대로 시작할건지를 설정

map_university.choropleth(geo_data=geo_str,
              data = un_region['학교명'],
              columns=[un_region.index, un_region['학교명']],
              fill_color='PuRd',
              key_on='feature.id')

#마커 설정
for i in range(data_size):

    folium.Marker(list(df_un.iloc[i][['위도', '경도']]), # 엑셀파일의 위도, 경도 위치에 맞게 마커를 찍어주도록 설정
                 popup = df_un.iloc[i][['학교명']], # 학교 명이 뜨도록 설정
                 icon = folium.Icon(color = 'black')).add_to(map_university) # 색상을 블랙으로 설정

map_university
map_university.save('university.html') #파일 html로 저장
```
